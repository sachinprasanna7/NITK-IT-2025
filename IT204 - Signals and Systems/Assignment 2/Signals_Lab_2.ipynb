```python
from numpy.fft import fft
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
sr=1000
t=np.linspace(0,1,1000)
x=signal.square(np.sin(2*np.pi*50*t))
X = fft(x)
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T
plt.subplot(121)
plt.plot(t,x)
plt.xlim(0,0.2)
plt.subplot(122)
plt.stem(freq, np.abs(X), 'b',markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('FFT Amplitude')
```
