```python
import matplotlib.pyplot as plt   #importing matplolib library
import numpy as np

# Numpy's arange() function has three arguments: start, stop, step. We start at zero, stop at 10 and step by 0.1 radians. 

x = np.arange(0,10*np.pi,0.1) 
y=np.cos(x)
plt.title("cosx")
plt.grid()
plt.axhline(y=0,color='k')
plt.plot(x,y)  #plot the graph
plt.show()     #display the graph

```

(c) Sawtooth function

```python
import matplotlib.pyplot as plt   #importing matplolib library
import numpy as np                #importing numpy as np
from scipy import signal


#use linespace:The NumPy linspace function is a tool in Python for creating numeric sequences that return evenly spaced numbers over a specified interval.
#It has start stop and number of plots.

x=np.linspace(0,1,1000,endpoint=True)
plt.grid()
plt.axhline(y=0,color='k')
plt.title("Sawtooth")

y=signal.sawtooth(2*np.pi*5*x)
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.plot(x,y)
plt.show()
```

(d) Square Function




```python
import matplotlib.pyplot as plt  
import numpy as np               
from scipy import signal


x=np.linspace(0,1,1000,endpoint=True)
y=signal.square(2*np.pi*5*x)
plt.grid()
plt.axhline(y=0,color='k')
plt.title("Square")
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.plot(x,y)
plt.show()
```

(e) Triangle Function

```python
import matplotlib.pyplot as plt   
import numpy as np                
from scipy import signal

x=np.linspace(0,1,1000,endpoint=True)
y=signal.sawtooth(2*np.pi*5*x,0.5)
plt.grid()
plt.axhline(y=0,color='k')

plt.title("Triangle")
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.plot(x,y)
plt.show()

```

**2.**
a)Impulse Function

```python
import matplotlib.pyplot as plt   
import numpy as np

x=np.arange(-10,10)
y=[]
for i in x:
  if i==0:
    y.append(1)
  else:
    y.append(0)
plt.stem(x,y, use_line_collection=True)
plt.grid()
plt.axhline(y=0, color="black")
plt.title("Impulse Function")
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.show()

```

(b) Step Function

```python
import matplotlib.pyplot as plt   #importing matplolib library
import numpy as np

x=np.arange(-10,10)
y=[]
for i in x:
  if (i<0):
    y.append(0)
  elif (i>0):
    y.append(1)
  else:
    y.append(0.5)  

plt.stem(x,y,use_line_collection="True")
plt.grid()
plt.axhline(y=0, color="black")
plt.title("Impulse Function")
plt.ylabel("Amplitude")
plt.xlabel("Time")

plt.show()
```

(c) Ramp Signal

```python
import matplotlib.pyplot as plt   #importing matplolib library
import numpy as np

x=np.arange(-10,10)
y=[]
for i in x:
  if (i<=0):
    y.append(0)
  elif (i>0):
    y.append(i)
    

plt.stem(x,y,use_line_collection="True")
plt.grid()
plt.axhline(y=0, color="black")
plt.title("Ramp Function")
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.show()
```

(d)Exponential Growing Signal

```python
import matplotlib.pyplot as plt   #importing matplolib library
import numpy as np

x=np.arange(-10,10)
y=[]
for i in x:
    y.append(np.exp(i))
    

plt.stem(x,y,use_line_collection="True")
plt.grid()
plt.axhline(y=0, color="black")
plt.title("Exponential Growth")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()


```

(e)Exponential Decaying Signal

```python
t = np.linspace(-10, 10, 35) 
y5 = []
for i in t:
  if(i >= 0):
    y5.append(np.exp(-i))
  else:
    y5.append(0)
    
plt.stem(t, y5, use_line_collection=True)
plt.title('(e) Exponentially Decaying')
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.axhline(y=0, color="black")
plt.grid()
plt.show()


```

**3**
a) Sine Signal

```python
import matplotlib.pyplot as plt   #importing matplolib library
import numpy as np

 

x1 = np.arange(0,10*np.pi,0.1) 
y1=np.sin(x1)


x2=np.arange(0,10*np.pi,0.5) 

y2=[]
for i in x2:
  y2.append(np.sin(i))

plt.plot(x1,y1)
plt.stem(x2,y2,use_line_collection="True")
plt.grid()
plt.axhline(y=0, color="black")
plt.title("Sine Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()


```

(b) cos(2pit)

```python
import matplotlib.pyplot as plt   
import numpy as np


x1 = np.linspace(0, 7, 500, endpoint = True) 
y1=np.cos(2*(np.pi)*x1)
x2=np.linspace(0, 7, 90, endpoint = True)  
y2=[]
for i in x2:
  y2.append(np.cos(2*(np.pi)*i))
plt.plot(x1,y1)
plt.stem(x2,y2,use_line_collection="True")
plt.grid()
plt.axhline(y=0, color="black")
plt.title("Cos(2*pi*t) Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

(c) y=cos(5t + sin (2t))


```python
import matplotlib.pyplot as plt  
import numpy as np


x1 = np.linspace(0, 7, 500, endpoint = True) 
y1=np.cos(5*(x1)+np.sin(2*(x1)))

x2=np.linspace(0, 7, 90, endpoint = True)  
y2=(np.cos(5*x2+np.sin(2*x2)))

plt.plot(x1,y1)
plt.stem(x2,y2,use_line_collection="True")
plt.grid()
plt.axhline(y=0, color="black")
plt.title("cos(5t+sin(2t)) Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

**4**(a) u[n] = x[n]+y[n]

```python
import matplotlib.pyplot as plt   
import numpy as np

t = np.arange(-3, 4, 1)
x = np.array([3, -2, 0, 1, 4, 5, 2])
y = np.array([0, 7, 1, -3, 4, 9, -2])
u = np.add(x, y)
plt.stem(t, u, use_line_collection=True)
plt.grid()
plt.title("x[n] + y[n]")
plt.ylabel('u')
plt.axhline(y = 0, color = 'black')
plt.xlabel('t')
plt.show()
```

(b) v[n]=x[n].w[n]

```python
import matplotlib.pyplot as plt   
import numpy as np
t = np.arange(-3, 4, 1)
x = np.array([3, -2, 0, 1, 4, 5, 2])
w = np.array([-5, 4, 3, 6, -5, 0, 1])
v = np.multiply(x, w)


plt.stem(t, v,use_line_collection=True)
plt.grid()
plt.title("x[n].w[n]")
plt.ylabel('v')
plt.axhline(y = 0, color = 'k')
plt.xlabel('t')
plt.show()

```

(c) s[n] = y[n]  w[n]

```python
import matplotlib.pyplot as plt   
import numpy as np

t = np.arange(-3, 4, 1)
y = np.array([0, 7, 1, -3, 4, 9, -2])
w = np.array([-5, 4, 3, 6, -5, 0, 1])
v = np.subtract(y, w)

plt.stem(t, v,use_line_collection=True)
plt.grid()
plt.title("s[n] = y[n]  w[n]")
plt.ylabel('s')
plt.axhline(y = 0, color = 'k')
plt.xlabel('t')
plt.show()
```

(d) r[n] = 4.5y[n]

```python
import matplotlib.pyplot as plt   
import numpy as np

t = np.arange(-3, 4, 1)
y = np.array([0, 7, 1, -3, 4, 9, -2])
r = np.multiply(4.5, y)
plt.stem(t, r,use_line_collection=True)
plt.grid()
plt.title('r[n] = 4.5y[n]')
plt.ylabel('r')
plt.axhline(y = 0, color = 'black')
plt.xlabel('t')
plt.show()
```
