```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
p = list(eval(input("Enter Search Pattern : ")))
sig = list(eval(input("Enter Input Signal : ")))
```

```python
plt.tight_layout()
plt.subplot(121)
plt.title("Search Pattern")
plt.plot(p)
plt.subplot(122)
plt.title("Signal")
plt.plot(sig)
```

```python
corr = signal.correlate(p,sig)
plt.plot(corr)
```

```python
max_vals = max(corr)
max_args = np.where(corr == max_vals)[0]
max_args = [i - int(len(p)/2) for i in max_args]
print(max_args)
```

```python
#Noisy input
sig = sig + np.random.normal(0,0.1,len(sig));
plt.tight_layout()
plt.subplot(121)
plt.title("Search Pattern")
plt.plot(p)
plt.subplot(122)
plt.title("Noisy Signal")
plt.plot(sig)
```

```python
corr = signal.correlate(p,sig)
plt.plot(corr)
max_vals = max(corr)
max_args = np.where(corr == max_vals)[0]
max_args = [i - int(len(p)/2) for i in max_args]
print(max_args)
```
