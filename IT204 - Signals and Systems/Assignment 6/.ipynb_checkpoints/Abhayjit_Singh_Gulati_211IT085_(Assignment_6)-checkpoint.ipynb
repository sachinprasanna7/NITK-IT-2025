Simple Input Signal


Single Matching

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

input_signal = np.array([6, 8, 6, 6, 8, 10, 8, 6, 4 ,3, 2, 4, 4, 4, 4, 4, 4, 4, 1, 1,1, 3, 1, 2, 4, 4, 4, 4, 4, 4, 4, 4, 4, 6, 8, 10, 8, 6, 4, 4])
search_signal= np.array([6, 8, 10, 8, 6])

match = signal.correlate(input_signal, search_signal, mode = 'same') #performing correlation between input signal and search signal

peak = np.max(match)
positions = []
for i in range(0, len(match)):
    if match[i] == peak:
        positions.append(i - 1)
        if len(positions)==1:   #to perform single matching
          break             

# Plot the input signal         
plt.plot(input_signal)

#Label the x-axis, y-axis, title
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Input Signal')
plt.grid(True)
plt.show()
print('\n')
# Plot the search signal  
plt.plot(search_signal)

#Label the x-axis, y-axis, title
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Search Pattern')
plt.grid(True)
plt.show()
print('\n')
# Plot the match signal  
plt.plot(match)

#Label the x-axis, y-axis, title
plt.xlabel('Position')
plt.ylabel('Correlation Amplitude')
plt.title('Correlation')
plt.grid(True)
plt.show()
print('\n')
# Displaying the final output
print('Pattern found at starting positions: ', positions)
```

Multi-Match

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

input = np.array([6, 8, 6, 6, 8, 10, 8, 6, 4 ,3, 2, 4, 4, 4, 4, 4, 4, 4, 1, 1,1, 3, 1, 2, 4, 4, 4, 4, 4, 4, 4, 4, 4, 6, 8, 10, 8, 6, 4, 4])
search = np.array([6, 8, 10, 8, 6])
match = signal.correlate(input, search, mode = 'same')    #performing correlation between input signal and search signal
peak = np.max(match)
positions = []
for i in range(0, len(match)):
  if match[i] == peak:
    positions.append(i - 1)          

# Plot the input signal
plt.plot(input)

#Label the x-axis, y-axis, title
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Input Signal')
plt.grid(True)
plt.show()
print('\n')

# Plot the search signal 
plt.plot(search)

#Label the x-axis, y-axis, title
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Search Pattern')
plt.grid(True)
plt.show()
print('\n')
# Plot the match signal 
plt.plot(match)

#Label the x-axis, y-axis, title
plt.xlabel('Position')
plt.ylabel('Correlation Amplitude')
plt.title('Correlation')

plt.grid(True)
plt.show()
print('\n')
# Displaying the final output
print('Pattern found at starting positions: ', positions)
```

Noisy Signal

Single Matching

```python
input = np.array([6, 8, 6, 6, 8, 10, 8, 6, 4 ,3, 2, 4, 4, 4, 4, 4, 4, 4, 1,1,1, 3, 1, 2, 4, 4, 4, 4, 4, 4, 4, 4, 4, 6, 8, 10, 8, 6, 4, 4])
search = np.array([6, 8, 10, 8, 6])

input = input + np.random.normal(0, 0.1, len(input)) #adding noise

match = signal.correlate(input, search, mode = 'same')      #performing correlation between input signal and search signal

peak = np.max(match)
positions = []

for i in range(0, len(match)):
  if abs(match[i] - peak) < 5:
    positions.append(i - 1)
    if len(positions)==1:           # to perform single matching
          break

# Plot the input signal    
plt.plot(input)

#Label the x-axis, y-axis, title
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Input Signal')
plt.grid(True)
plt.show()
print('\n')

# Plot the search signal 
plt.plot(search)

#Label the x-axis, y-axis, title
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Search Pattern')
plt.grid(True)
plt.show()
print('\n')

# Plot the match signal 
plt.plot(match)
#Label the x-axis, y-axis, title
plt.xlabel('Position')
plt.ylabel('Correlation Amplitude')
plt.title('Correlation')
plt.grid(True)
plt.show()
print('\n')
print('Pattern found at starting location: ', positions)
```
